### Step 9. Create a function called majority that return a boolean value to a new column called legal_drinker

```python
returnVar = True
def majority(x):
    if (x > 17):
        returnVar = True
    else:
        returnVar = False
    return returnVar    
```

```python
majority(17)
```

```python
dfNext['legal_drinker'] = dfNext['age'].apply(lambda x: majority(x))
```
